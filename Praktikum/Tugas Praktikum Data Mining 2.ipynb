```python
import numpy as np

data_numerik = df.select_dtypes(include=['number'])
QI = np.percentile(df, 25)
Q3 = np.percentile(df, 75)
IQR = Q3 - QI

lower_bound = QI - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers= np.where((df < lower_bound)|(df>upper_bound))
print( "Indeksoutliers: " ,outliers)
```

```python
jumlah =df['brand_name'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))
plt.bar(jumlah.index, jumlah,width=0.7)
plt.xlabel('Brand')
plt.ylabel('Jumlah Brand')
plt.title('Brand')
plt.show()

```
