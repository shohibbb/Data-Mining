```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/drive/MyDrive/Semester 6/Data Mining/smartphones.csv')
```

```python
df.info()
```

# Data Cleaning

```python
df.isna().sum()
```

```python
df.head()
```

Mengisi kolom avg_rating

```python
df.avg_rating.fillna(value = df.avg_rating.mean(), inplace = True)
```

Mengisi kolom Processor_brand

```python
df.processor_brand.fillna(value = 'unkwon', inplace = True)
```

Mengisi kolom mun_cores

```python
df.num_cores.fillna(value = df.num_cores.mode()[0], inplace = True)
```

Mengisi kolom processor_speed

```python
df.processor_speed.fillna(value = df.processor_speed.mean(), inplace = True)
```

Mengisi kolom battery_capacity

```python
df.battery_capacity.fillna(value = df.battery_capacity.median(), inplace = True)
```

Mengisi kolom fast_charging

```python
df.fast_charging.fillna(value = df.fast_charging.mode()[0], inplace = True)
```

Mengisi kolom os

```python
df['os'].value_counts()
```

```python
df.os.fillna('other', inplace = True)
```

Mengisi kolom primary_camera_front

```python
df.primary_camera_front.fillna(df['primary_camera_front'].median(), inplace = True)
```

```python
df.isna().sum()
```

# Encoding

```python
labelencoder = LabelEncoder()
```

```python
df['processor_brand'].value_counts()
```

```python
df['processor_brand'] = labelencoder.fit_transform(df['processor_brand'])
```

```python
df['os'].value_counts()
```

```python
df['os'] = labelencoder.fit_transform(df['os'])
```

```python
df.tail()
```

# Normalisasi

```python
scaler = MinMaxScaler()
```

```python
df['battery_capacity'] = scaler.fit_transform(df['battery_capacity'].values.reshape(-1, 1))
```

```python
df['internal_memory'] = scaler.fit_transform(df['internal_memory'].values.reshape(-1, 1))
```

```python
df['resolution_height'] = scaler.fit_transform(df['resolution_height'].values.reshape(-1, 1))
```

```python
df['resolution_width'] = scaler.fit_transform(df['resolution_width'].values.reshape(-1, 1))
```

# Data Splitting


```python
df_split = df.drop(["brand_name","model"], axis=1)
```

```python
features = df_split.drop(["avg_rating"], axis=1)
target = df["avg_rating"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

# Data reduction

```python
pca = PCA(n_components=0.95)
```

```python
pca.fit(features)
```

```python
n_components = pca.n_components_
```

```python
features_reduced = pca.transform(features)
```
