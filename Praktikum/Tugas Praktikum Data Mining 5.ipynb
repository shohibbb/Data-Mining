```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 20))

for i, col in enumerate(important_columns):
    row, col_num = divmod(i, 4)
    sns.histplot(df[col], bins=5, kde=False, ax=axes[row, col_num])
    axes[row, col_num].set_title(col)

    # percentage
    total = len(df[col])
    for p in axes[row, col_num].patches:
        height = p.get_height()
        axes[row, col_num].text(p.get_x() + p.get_width() / 2.,
                                height + 0.01 * total,
                                '{:.2%}'.format(height / total),
                                ha="center")

plt.tight_layout()
plt.show()
```

## Modeling

### using scale

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()

after_scaled = sc.fit_transform(df)

df_scaled = pd.DataFrame(data=after_scaled, columns=df.columns)
df_scaled.head(10)
```

```python
from sklearn.decomposition import PCA

pca_scaled = PCA(n_components=None)
dfx_pca_scaled = pca_scaled.fit(df_scaled)
dfx_pca_scaled.explained_variance_ratio_

n_components = sum(dfx_pca_scaled.explained_variance_ratio_ > 0.1)
pca_scaled = PCA(n_components=n_components)
after_PCA_scaled = pca_scaled.fit_transform(df_scaled)
print(n_components)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


wcss_scaled = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(after_PCA_scaled)
    wcss_scaled.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_scaled)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km_scaled = KMeans(n_clusters=3)
y_pred_scaled = km_scaled.fit_predict(after_PCA_scaled)
```

```python
plt.scatter(after_PCA_scaled[:, 0], after_PCA_scaled[:, 1], c=km_scaled.labels_, cmap="viridis")
plt.scatter(km_scaled.cluster_centers_[:, 0], km_scaled.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title(f"K-means Clustering (K={3})")
plt.show()
```

### without scaled

```python
pca = PCA(n_components=None)
dfx_pca = pca.fit(df)
dfx_pca.explained_variance_ratio_

n_components = sum(dfx_pca.explained_variance_ratio_ > 0.1)
pca = PCA(n_components=n_components)
after_PCA = pca.fit_transform(df)
print(n_components)
```

```python
wcss = []
for i in range(1, 11):
    # kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    #for the sake of simplicity, you can just specify the number of cluster
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(after_PCA)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(after_PCA)
```

```python
plt.scatter(after_PCA[:, 1], after_PCA[:, 3], c=km.labels_, cmap="viridis")
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=300, c='red')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title(f"K-means Clustering (K={3})")
plt.show()
```
